```python
#import requests
from urllib.request import urlopen
```

```python
url = "https://www.nationalgeographic.com/animals/mammals/facts/domestic-dog"
```

```python
page = urlopen(url)
```

```python
page
```

```python
html_bytes = page.read()

html = html_bytes.decode("UTF-8")
```

```python
print(html)
```

```python
# To find a title

title_index = html.find("<title>")

start_index = title_index + len("<title>")
start_index
```

```python
# Getting the index of the closing tag

end_index = html.find("</title>")
end_index
```

```python
# Extracting the title by slicing

title = html[start_index:end_index]
title
```

```python
# now the above method can be used in simpler html documents only
# Following is a more reliable way

import re

pattern = "<title.*?>.*?</title.*?>"
match_results = re.search(pattern, html, re.IGNORECASE)
match_results.group()

title = match_results.group()
```

```python
# Remove HTML tags

title = re.sub("<.*?>", "", title) 
print(title)
```

# Using Beautiful Soup

```python
from bs4 import BeautifulSoup  
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Dog"
```

```python
# Make a GET request to fetch the raw HTML content

page_content = requests.get(url).text
```

```python
# Parse the html content
# html5lib is a pure-python library for parsing HTML.

soup = BeautifulSoup(page_content, "html5lib")
soup.prettify()
```

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
soup.title.text
```

### Getting all the links in the document

##### Here we use soup.find_all('a') because all the links are in the <a> tags

```python
for link in soup.find_all('a'):
  print(link.get('href'))
```

### Getting all the text in the page

```python
print(soup.get_text())
```

```python
print(soup.get_attribute_list)
```

```python
images = soup.select('img') 

images_url = images[0]['src'] 

print(images_url)
```

```python
for i in soup.find_all('img'):
  print(i.get('src'))
```

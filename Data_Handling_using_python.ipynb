##SYSTEMATIC SAMPLING

```python
# Define systematic sampling function
def systematic_sampling(df, step):
 
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
 
 
# Obtain a systematic sample and save it in a new variable
systematic_sample = systematic_sampling(df_split, 3)
 
# View sampled data frame
display(systematic_sample)
```

# DIMENSIONALITY REDUCTION

```python
from IPython.display import IFrame

'''When this object is returned by an input cell or passed to the display function, it will result 
in Visual controls being displayed in the frontend (only works in the notebook)'''

documentation = IFrame(src='https://www.javatpoint.com/dimensionality-reduction-technique', width=1000, height=450)
display(documentation)
```

Dimensionality reduction is always applied on a standardized data.

 All the variables should be on the same scale before applying PCA, otherwise, a feature with large values will dominate the result.

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df_split['binned'] = encoder.fit_transform(df_split['binned'])
```

```python
a = df_split.iloc[:,0:101]
b = df_split.iloc[:,101:102]
```

```python
# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
a = scaler.fit_transform(a)
```

```python
# Visualize

from sklearn.decomposition import PCA

pca = PCA()
pca.fit_transform(a)
pca_variance = pca.explained_variance_

plt.figure(figsize=(8, 6))
plt.bar(range(101), pca_variance, alpha=0.5, align='center', label='individual variance')
plt.legend()
plt.ylabel('Variance ratio')
plt.xlabel('Principal components')
plt.show()
```

We used PCA to get the list of features and plot which features have the most explanatory power, or have the most variance. These are the principle components. It looks like around 10-15 of the features explain the majority, almost 95% of our data

```python
pca2 = PCA(n_components=15)
pca2.fit(a)
x_3d = pca2.transform(a)

plt.figure(figsize=(8,6))
plt.scatter(x_3d[:,0], x_3d[:,5], c=df_split['binned'])
plt.show()
```

```python
pca3 = PCA(n_components=2)
pca3.fit(a)
x_3d = pca3.transform(a)

plt.figure(figsize=(8,6))
plt.scatter(x_3d[:,0], x_3d[:,1], c=df_split['binned'])
plt.show()
```
